# Handling Missing values

# MSZoning
have 4 missing values and its an object ( string ) with propably 3-4 category. So we will fill that with the most frequent cateogory or simply saying mode

```python
mode_value = test_data['MSZoning'].mode()[0]
test_data['MSZoning'].fillna(mode_value, inplace=True)
```

# LotFrontage, MasVnrArea
 data type is float. So filling the null values with their mean

```python
# lotfrontage
mean_value = test_data['LotFrontage'].mean()
test_data['LotFrontage'].fillna(mean_value, inplace=True)

# MasVnrArea
test_data['MasVnrArea'].fillna(test_data['MasVnrArea'].mean(), inplace=True)
```

# Alley, PoolQC, Fence, MiscFeature
have more than 80% missing values so drop it

```python
columns_to_drop = ['Alley','PoolQC','Fence','MiscFeature']

#Drop columns
test_data  = test_data.drop(columns=columns_to_drop)
```

```python
category=np.array(['MasVnrType', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 
                   'BsmtFinType2', 'FireplaceQu','GarageType','GarageYrBlt',
                   'GarageFinish', 'GarageQual', 'GarageCond' ])


for element in category:
    test_data[element].fillna(test_data[element].mode()[0], inplace=True)
```

```python
category2=['SaleType', 'GarageArea', 'GarageCars', 'Functional', 'KitchenQual',
 'BsmtHalfBath', 'BsmtFullBath','BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', 
'BsmtFinSF1','Utilities','Exterior1st', 'Exterior2nd']


for elem in category2:
    test_data[elem].fillna(test_data[elem].mode()[0], inplace=True)
```
